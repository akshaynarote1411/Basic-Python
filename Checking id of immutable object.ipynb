```python
x=1
print(x)
print(id(x))

while x<5:
    print(x)
    print(id(x))
    x=x+1
    
print(x)
print(id(x))
```

```python
x=1
print(x)
print(id(x))

l = []
print(id(l))

while x<5:
    
    l.append(x)
    print("id of list :",id(l))   # It will same for every item(number) in iteration cause list is mutable object. 
    
    print(x)
    print("id of each number in iteration:",id(x))   # It will be different for every item(number) in iteration cause items are int data types and it is immutable.
    x=x+1
    
print(x)
print(id(x))
```
