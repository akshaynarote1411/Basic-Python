```python
# loops on dictionary datatype work as same but we have to write it differently, we can't directly write the dict variable like other datatype variable

d = {'a':101, 'b':200, 'c':500}
print(type(d))

for i in d:
    print(d[i])
    if (d[i] % 2) == 0:
        print(f"{d[i]} is even number")
    
    else : 
        print(f"{d[i]} is odd number")
```

```python
d = {'a':101, 'b':200, 'c':500}
print(type(d))

for i in d:
    print(i)
    print(d[i])
```

```python
d = {'a':101, 'b':200, 'c':500}
print(type(d))

for i in d.keys():
    print(i)
```

```python
d = {'a':101, 'b':200, 'c':500}
print(type(d))

for i in d.values():
    print(i)
```

```python
d = {'a':101, 'b':200, 'c':500}
print(type(d))

for i in d.items():
    print(i)
```
