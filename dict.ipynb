```python
Names = ["Ramesh","Suresh","Satish"]
Age = [20,22,24]

# I want to print like : Ramesh has age 20, Suresh has age 22, Satish has age 24

for i in range(len(Names)):
    print(Names[i],"has age",Age[i])
     # or print(f"{Names[i]} has age {Age[i]}")
```

```python
dict = {"ram":1,"sam":2,"tom":3}
dict[0]   # Indexing is not allowed in dictionaries because in key-value pair key is important not value, so we can't access by giving index number
```

```python
for i in dict:
    print(i)        # It will gives only keys in output from dictionaries's items not values, so we can access the item by giving keys as the index not index number.
```

```python
dict["ram"]    # It will gives value of passing key from dict, so we have to access the values from the dict by passing keys of items.
```
