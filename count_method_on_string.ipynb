#### find the count of the substring from the string without using the count method?

```python
str1 = "ola ola ola"

count = 0

for i in range(len(str1)):
    print(i,str1[i:i+3])   # This is used to slice the the three letters from the str1 to compare with 'ola'
    if str1[i:i+3]=="ola":
        count=count+1
count


# This program will work as below
# for i=0, str1[0:0+3], str1[0:3]-> 'ola'   -> count=1
# for i=1, str1[1:1+3], str1[1:4]-> 'la '   -> count=1
# for i=2, str1[2:2+3], str1[2:5]-> 'a o'   -> count=1
# for i=3, str1[3:3+3], str1[3:6]-> ' ol'   -> count=1
# for i=0, str1[4:4+3], str1[4:7]-> 'ola'   -> count=2
.....
```
