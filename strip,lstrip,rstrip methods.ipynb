```python
s = "python "

s.strip()    # It is used to remove the unwanted substring from the string from both sides.   
```

```python
s1 = "python***"
s1.strip("*")
```

```python
s2 = "  python  "
s2.strip(" ")
```

```python
s3 = "  python***"
s3.strip(" *")
```

```python
s3.lstrip()  # It is used to remove the unwanted substring from the string from left side.
```

```python
s2.lstrip()
```

```python
s3.rstrip("*")   # It is used to remove the unwanted substring from the string from right side.
```

```python
s2.rstrip()
```
