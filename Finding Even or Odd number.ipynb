```python
print("Enter the number, when completed please enter #: ")

l1 = []
l2 = []

for i in range(100):
    num  = input("Enter the number : ")
    
    if num == '#':
        break
    elif (int(num) % 2)== 0:
        l1.append(int(num))
    else:
        l2.append(int(num))
        
print("Even Numbers that you have inputed: ",l1)
print("Odd Numbers that you have inputed: ",l2)
```

```python
print("Enter the numbers , when completed pleas enter # :")

l1 = []
l2 = []
s_even = s_odd = 0

for i in range(100):
    num = input("Enter the number : ")
    
    if num == '#':
        break
        
    elif (int(num)%2) == 0:
        l1.append(int(num))
        s_even = s_even + int(num)
        
    else:
        l2.append(int(num))
        s_odd = s_odd + int(num)
        
print(f"list of Even Numbers : {l1} & sum of even numbers : {s_even}")
print(f"list of Odd Numbers : {l2} & sum of odd numbers : {s_odd}")
```
