#### case-2 : for loop with if condition

```python
 # case-1 : [<output> <for loop>]

 # case-2 : [<output> <for loop> <if cond>]
```

```python
l = list(range(1,30))

evens = []

for i in l:
    if i%2==0:
        evens.append(i)
        
evens
```

```python
[i for i in l if i%2==0]
```

```python
output = [i for i in Input if "#" in i]
output
```

#### case-3 : for loop with if else conditions

```python
# case-1 : [<output> <for loop>]
# case-2 : [<output> <for loop> <if cond>]
# case-3 : [<if output> <if cond> else <else output> <for loop>]   # while coding in list comprehensions if "if cond" is there then immediately after "if cond" write "else" after that "else output". 
```

```python
l = [1000,2000,3000,550,200,1770]
o = []
# if value is greater than 1000 then expensive else offerdable

for i in l:
    if i>=1000:
       o.append("expensive")
    
    else:
        o.append("offerdable")
        
o
```

```python
o = ["expensive" if i>=1000 else "offerdable" for i in l]
o
```

#### case-4 : for loop with if elif conditions 

note: there is no if elif conditions in list comprehensions in python, but by using case-3 we can generate the case-4 as bellow

```python
# case-1 : [<output> <for loop>]
# case-2 : [<output> <for loop> <if cond>]
# case-3 : [<if output> <if cond> else <else output> <for loop>]
# case-4 : [<if out> <if cond> else <elif out> <if cond> else <elif out> <if cond> else <else out>]
```

```python
l = [70,10,90,77,50,60]

out=[]

for i in l:
    if i>=90:
        out.append("A")
        
    elif i>=70:
        out.append("B")
        
    elif i>=50:
        out.append("C")
        
    else:
        out.append("F")
        
        
out
```

```python
out = ["A" if i >=90 else "B" if i>=70 else "C" if i>=50 else "F" for i in l]
out
```

### Valrus Operator(:=):
 In list comprehensions '=' operator will not work, so instead of it we have to use Valrus Oerator(:=) to execute the operations in list comprehensions


```python
s=0
s = [s:=s+i for i in range(1,11)]    # This is Called as Valrus Operater in which we use ':=' or '==' instead of '=' to operator in list comprehensions.
s
```

```python
marks = [45,76,89,53,67,90,82,49,33,99]

# filter students who passed(marks>=50)

passed = [f"{i} passed" for i in marks if i>=50]
passed
```

```python
# Calculate the grades for all students

[f"{i}-> A" if i>=85 else f"{i}-> B" if i>=70 else f"{i}-> C" if i>=50 else f"{i}-> F" for i in marks]
```

```python
# Add grace marks if student failed

[f"{i}->{i+5}" for i in marks if i<50]
```

```python
# Calculate the average percentage of passed students


sum = 0
[sum:=sum+i for i in marks if i>=50]
avg = sum/len(passed)
avg
```
